```python
# V4 Error handling


# import required modules 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
import time


def gym_bot(day,slot):
    
    booking_xpath = "/html/body/div[4]/div[1]/section/div/div/div/div/ul/li[" + str(day) + "]/ul/li[" + str(slot) + "]/div[6]/a"
    

    # Web driver settings for WayScript
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    driver = webdriver.Chrome(options = options)


    driver.get("GYM URL")
    delay = 30 # seconds
    #Select dropdown menu
    
    

    # deselect gym
    element = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[4]/header/div[1]/div[2]/div[2]/div[1]/ul/li[2]/a/i')))
    driver.execute_script("arguments[0].click();", element)
    # deselect gym
    element = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[4]/header/div[1]/div[2]/div[2]/div[1]/ul/li[1]/a/i')))
    driver.execute_script("arguments[0].click();", element)

    #deselect gym
    element = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[4]/header/div[1]/div[2]/div[2]/div[1]/ul/li[1]/a/i')))
    driver.execute_script("arguments[0].click();", element)
    time.sleep(1)

    #select labbet
    select = Select(driver.find_element_by_xpath('//*[@id="unit"]'))
    select.select_by_index(3)
    time.sleep(1)

    # click search button to filter
    element = driver.find_element_by_xpath('//*[@id="search-selection"]')
    driver.execute_script("arguments[0].click();", element)
    time.sleep(2)

    # #select workout time 1615
    element = driver.find_element_by_xpath(booking_xpath)
    driver.execute_script("arguments[0].click();", element)


    time.sleep(1)
    # enter email
    driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div[1]/div/form/div/ul/li[5]/input').send_keys('xxx@gmail.com')
    time.sleep(1)
    #enter password
    driver.find_element_by_xpath('//*[@id="Password"]').send_keys('PASSWORD')
    time.sleep(1)
    # #click book pass
    element = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div[1]/div/form/div/input')
    driver.execute_script("arguments[0].click();", element)
    time.sleep(15)
    
    #close confirmation window to confirm that the booking has worked
    element = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div[1]/div[1]/button')))
    driver.execute_script("arguments[0].click();", element)
    
    
    driver.quit()
    
# Do a while loop with a count that will restart the script if an error occurs for x attempts.

attempts = 0
while attempts < 3:
    try: 
        #select the day (1-7) and slot (6 is 1700-1930)
        gym_bot(3,7)
        attempts = 6
        
    except Exception as e:
        attempts += 1

if attempts == 6:
    print('Booking Completed')

else:
    print('Booking Failed')
        
    
```
